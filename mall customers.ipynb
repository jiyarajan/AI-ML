```python
sns.kdeplot(df['Age'], fill=True, color='skyblue')
plt.title('Age Distribution (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
sns.histplot(data=df, x='Age',bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
gender = df['Gender'].replace({0:'Male',1:'Female'}).value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()

print("Gender Counts:\n", gender_counts)
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.show()
```
